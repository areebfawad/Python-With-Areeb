```python
def sports_details():
    
    flag = True
    data = {}
    while flag:
        name = input('Enter name: ').title()
        sports = input('Enter sports: ').title()
        if name == 'Q' or sports == 'Q':
            flag = False
        else:
            if name not in data:
                data[name] = [sports]
            else:
                data[name].append(sports)
    with open('testing1.txt', 'a') as f:
        for name, sport in data.items():
                f.write(f"{name}: {', '.join(sport)}\n")


```

```python
sports_details()
```
