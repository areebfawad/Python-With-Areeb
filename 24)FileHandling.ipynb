- r+:
    - file should exists
    - can use read and write both

- w+:
    - work similarily to normal write mode
    - can use read and write both

- a+:
    - work similarily to normal append mode
    - can use read and write both


# Context Manager(with)

```python
# f------alias
with open('testing1.txt') as f:
    content = f.read()
```

```python
content
```

```python
with open('newfile.txt', 'w+') as f:
    f.write('I am using context manager')
    f.seek(0)
    content = f.read()
```

```python
content
```

```python
with open('newfile.txt', 'w') as f:
    f.write('I am using context manager')
    f.seek(0)
    content = f.read()
```

```python
with open('testing1.txt', 'a+') as f:
    f.seek(0)
    content = f.read()
```

```python
content
```

```python
with open('testing1.txt', 'a+') as f:
    f.write(' ALI KHAN')
    f.seek(0)
    content = f.read()
    
```
