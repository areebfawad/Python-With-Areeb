```python
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe(self):
        print(f'{self.restaurant_name}...{self.cuisine_type}')
        
    def open(self):
        print(f'This restaurant {self.restaurant_name} is open')
```

# Modifying Attribute Values


```python
class User:
    def __init__(self, first_name, last_name, gender, age):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = 'SMIT'

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')

    def set_attribute(self, new_campus):
        self.campus = new_campus
```

```python
kashan = User('Kashan', 'khan', "Male", 20)
yaman = User('Yaman', 'peshawari', "Male", 10)
ghafar = User('Ghafar', 'sawati', "Male", 12)
```

```python
#Modifying an Attributes Value Directly

kashan.campus = "JDC"
```

```python
kashan.campus
```

```python
yaman.campus
```

```python
# Modifying an Attributes Value Through a Method
kashan.set_attribute('JDC')
```

```python
kashan.campus
```

```python
yaman.campus
```

```python
class User:
    def __init__(self, first_name, last_name, gender, age):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = 'SMIT'

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')

    def set_attribute(self,attribute, value):
        # self.attribute = value
        setattr(self, attribute, value)



```
