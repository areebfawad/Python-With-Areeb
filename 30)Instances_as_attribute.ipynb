Base Class (ElectronicDevice): This is the parent class: 
Attribute: name, price, power_status  
Methods: (power_on()) that all electronic devices share.





Subclass (Laptop): This is the child class. It inherits (automatically receives) all the attributes and methods from ElectronicDevice.
Attributes:(screen_size_inches, ports)
Methods: list_ports()

```python
class ElectronicDevice:
    def __init__(self, name, price, power_status):
        self.name = name 
        self.price = price 
        self.power_status = power_status 

    def power_on(self):
        print('Power is on')
```

```python
class Laptop(ElectronicDevice):
    def __init__(self, name, price, power_status, screen_size, ports):
        super().__init__(name, price, power_status)
        self.screen_size = screen_size
        self.ports = ports

    def list_ports(self):
        print(f"These are the ports {', '.join(self.ports)}")
```

```python
laptop1 = Laptop('HP', 12000, True, 12.3, ['HDMI', 'USB'])
```

```python
laptop1.list_ports()
```

```python
laptop1.price
```

```python
class User:
    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age
    
    def describe_user(self):
        print(f'{self.fname} - {self.lname} - {self.age}')
    
    def greet_user(self):
        print(f'Hello, {self.fname} {self.lname}')
```

```python
class Admin(User):
    def __init__(self, fname, lname, age, privileges):
        super().__init__(fname, lname, age)
        self.privileges = privileges

    
    def show_privileges(self):
        for privilege in self.privileges:
            print(f"- {privilege}")
```

```python
ali = Admin('Ali', "khan", 12, ['Can Teach', 'Can Code'])
```

```python
ali.show_privileges()
```

# Instances as Attributes


```python
class Car:
    def __init__(self,make, model, year):
        # Attributes
        self.make = make
        self.model = model
        self.year =  year 
    
    # Methods
    def apply_accelerator(self):
        print('Accelerator is applied')
    
    def apply_breaks(self):
        print('Brakes are applied')
    
```

```python
class Battery:
    def __init__(self, company, voltage, capacity):
        self.company = company  
        self.voltage = voltage  
        self.capacity = capacity  

    def battery_info(self):
        print(f'Company: {self. company}\nVoltage:{self.voltage}\nCapacity: {self.capacity}')
```

```python
class ElectricCar(Car):
    def __init__(self,make, model, year, company, voltage, capacity):
        super().__init__(make,model,year)

        # Additional Attribute
        self.battery = Battery(company, voltage, capacity)


    # Additional Method
    def recharge(self):
        print('The battery is recharged now ')
    
```

```python
# b1 = Battery('osaka','12V', '55AH')
newec = ElectricCar('Honda', 'Civic', 2025,'osaka','12V', '55AH')
```

```python
newec.battery.battery_info()
newec.battery.company
```

```python
newec.apply_accelerator()
```
