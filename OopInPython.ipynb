## Encapsulation
- Binding of methods with attributes

```python
class User:
    def __init__(self, first_name, last_name, gender, age, campus = "SMIT"):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = campus

    def change_campus(self, new_campus):
        self.campus = new_campus

    def show_campus(self):
        return self.campus

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
```

```python
kashan = User('Kashan', 'khan', "Male", 20)
yaman = User('Yaman', 'peshawari', "Male", 10)
ghafar = User('Ghafar', 'sawati', "Male", 12)
```

```python
kashan.change_campus('JDC') # setter
```

```python
kashan.show_campus() #getter
```

```python
class User:
    def __init__(self, first_name, last_name, gender, age, campus = "SMIT"):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = campus

    def change_details(self, new_first_name='', new_last_name='', new_gender='', new_age=None, new_campus=''):
        if new_first_name:
            self.first_name = new_first_name
        if new_last_name:
            self.last_name = new_last_name
        if new_gender:
            self.gender = new_gender
        if new_age:
            self.age = new_age
        if new_campus:
            self.campus = new_campus

    def show_campus(self):
        return self.campus

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
```

```python
kashan = User('Kashan', 'khan', "Male", 20)
yaman = User('Yaman', 'peshawari', "Male", 10)
ghafar = User('Ghafar', 'sawati', "Male", 12)
```

```python
kashan.change_details('ali', 'khan')
```

```python
class User:
    def __init__(self, first_name, last_name, gender, age, campus = "SMIT"):
        self.first_name = first_name
        self.last_name = last_name
        self.gender = gender
        self.age = age
        self.campus = campus
        self.counter = 1
        self.update_object_data()

    def update_object_data(self):
        data = self.__dict__
        with open('testing1.txt', 'a+') as f:
            f.write(f'User{self.counter}:\n')
            for attr, value in data.items():
                self.counter += 1
                if '_' in attr:
                    attr = ' '.join(attr.split('_')).title()
                else:
                    attr = attr.title()
                f.write(f'{attr}: {value}\n')
            f.write('\n')

        

    def describe_user(self):
        print(f"User's name is {self.first_name} {self.last_name}\nAge: {self.age}\nGender: {self.gender}\nCampus: {self.campus}")

    def greet_user(self):
        print(f'Hello {self.first_name}Welcome to our page')
```
