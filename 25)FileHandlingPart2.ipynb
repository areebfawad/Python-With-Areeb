```python
def sports_details(name, sport):
    with open('testing1.txt', 'a') as f:
        f.write(f'{name.title()}: {sport.title()}')
    print(f'{name.title()}: {sport.title()}')
```

```python
sports_details('ali khan', 'cricket')
```

```python
def sports_details():

    flag = True
    while flag:
        name = input('Enter name: ')
        sports = input('Enter sports: ')
        if name == 'q' or sports == 'q':
            flag = False
        else:
            with open('testing1.txt', 'a') as f:
                f.write(f'{name.title()}: {sports.title()}\n')
            print(f'{name.title()}: {sports.title()}')
```

```python
sports_details()
```
