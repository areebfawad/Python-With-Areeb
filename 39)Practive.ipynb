```python
import pandas as pd
# Write a solution to calculate the bonus of each employee. The bonus of an employee is 100% of their salary if the ID of the employee is an odd number and the employee's name does not start with the character 'M'. The bonus of an employee is 0 otherwise.

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    result = (employees['employee_id'] % 2 != 0) &  (~ employees['name'].str.startswith('M'))
    employees['salary'] = employees['salary'].where(result, 0)
    employees.drop('name', axis=1, inplace=True)
    employees.rename(columns={'salary': 'bonus'}, inplace=True)
    employees.sort_values(by='employee_id', ascending=True, inplace=True)
    return employees
```

```python
import pandas as pd

def fix_names(users: pd.DataFrame) -> pd.DataFrame:
    users['name']  = users['name'].str.capitalize()
    users.sort_values(by='user_id', inplace=True)
    return users
```
